```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yolov5s_result = pd.read_csv("./paper_use_result/yolov5s_visdrone_gyd_epoch200/results.csv")
yolov5n_result = pd.read_csv("./paper_use_result/yolov5n_visdrone_gyd_epoch200/results.csv")
```

```python
epochn = yolov5n_result['epoch']
epochs = yolov5s_result['epoch']
yolov5s_train_obj_loss = yolov5s_result['train/obj_loss']
yolov5s_train_box_loss = yolov5s_result['train/box_loss']
yolov5s_val_obj_loss = yolov5s_result['val/obj_loss']
yolov5s_val_box_loss = yolov5s_result['val/box_loss']
yolov5s_precision = yolov5s_result['metrics/precision']
yolov5s_recall = yolov5s_result['metrics/recall']
yolov5s_mAP05 = yolov5s_result['metrics/mAP_0.5']
yolov5s_mAP05_95 = yolov5s_result['metrics/mAP_0.5:0.95']


yolov5n_train_obj_loss = yolov5n_result['train/obj_loss']
yolov5n_train_box_loss = yolov5n_result['train/box_loss']
yolov5n_val_obj_loss = yolov5n_result['val/obj_loss']
yolov5n_val_box_loss = yolov5n_result['val/box_loss']
yolov5n_precision = yolov5n_result['metrics/precision']
yolov5n_recall = yolov5n_result['metrics/recall']
yolov5n_mAP05 = yolov5n_result['metrics/mAP_0.5']
yolov5n_mAP05_95 = yolov5n_result['metrics/mAP_0.5:0.95']
```

```python
# train obj_loss
fig, ax = plt.subplots(2, 4)
fig.tight_layout()
plt.subplot(2, 4, 1)
plt.plot(epochs, yolov5s_train_obj_loss, marker='.', color='#F1B94A', linewidth=2, markersize=8)
plt.plot(epochn, yolov5n_train_obj_loss, linewidth=2, color='#0D75DC', markersize=8)
plt.title('train/obj_loss', fontsize=10)


# train obj_loss
plt.subplot(2, 4, 2)
plt.plot(epochs, yolov5s_train_box_loss, marker='.', color='#F1B94A', linewidth=2, markersize=8)
plt.plot(epochn, yolov5n_train_box_loss, linewidth=2, color='#0D75DC', markersize=8)
plt.title('train/box_loss', fontsize=10)


# val obj_loss
plt.subplot(2, 4, 3)
plt.plot(epochs, yolov5s_val_obj_loss, marker='.', color='#F1B94A', linewidth=2, markersize=8)
plt.plot(epochn, yolov5n_val_obj_loss, linewidth=2, color='#0D75DC', markersize=8)
plt.title('val/obj_loss', fontsize=10)


# train obj_loss
plt.subplot(2, 4, 4)
plt.plot(epochs, yolov5s_val_box_loss, marker='.', color='#F1B94A', linewidth=2, markersize=8)
plt.plot(epochn, yolov5n_val_box_loss, linewidth=2, color='#0D75DC', markersize=8)
plt.title('val/box_loss', fontsize=10)



plt.subplot(2, 4, 5)
plt.plot(epochs, yolov5s_precision, marker='.', color='#F1B94A', linewidth=2, markersize=8)
plt.plot(epochn, yolov5n_precision, linewidth=2, color='#0D75DC', markersize=8)
plt.title('metric/precisin', fontsize=10)


plt.subplot(2, 4, 6)
plt.plot(epochs, yolov5s_recall, marker='.', color='#F1B94A', linewidth=2, markersize=8)
plt.plot(epochn, yolov5n_recall, linewidth=2, color='#0D75DC', markersize=8)
plt.title('metric/recall', fontsize=10)


plt.subplot(2, 4, 7)
plt.plot(epochs, yolov5s_mAP05, marker='.', color='#F1B94A', linewidth=2, markersize=8)
plt.plot(epochn, yolov5n_mAP05, linewidth=2, color='#0D75DC', markersize=8)
plt.title('metric/mAP_0.5', fontsize=10)


plt.subplot(2, 4, 8)
plt.plot(epochs, yolov5s_mAP05_95, marker='.', color='#F1B94A', linewidth=2, markersize=8)
plt.plot(epochn, yolov5n_mAP05_95, linewidth=2, color='#0D75DC', markersize=8)
plt.title('metric/mAP_0.5:0.95', fontsize=10)
plt.legend(['yolov5s', 'yolov5n'], loc='lower right')

plt.savefig("visdrone2019_gyd.pdf")
plt.show()
```
